```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import numpy as np

import glob
```

```python
#Normalise data sets
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(min_max_scaler.fit_transform(pd.DataFrame(df[col])),columns=[col])
    return df
```

```python
path = "/media/joao/EXTERNAL_USB/1-BGL-ML_roots/data_with_JetMatching/"
all_Signal = glob.glob(path + "/314p99_*.csv")
all_SingleTop = glob.glob(path + "/SingleTop_Wtchannel_*.csv")
all_ttbar = glob.glob(path + "/ttbar_*_t.csv")
all_ttbarV = glob.glob(path + "/ttbar_V_*.csv")
all_WW = glob.glob(path + "/WW_*.csv")
all_ZZ = glob.glob(path + "/ZZ_*.csv")
all_ZW = glob.glob(path + "/ZW_*.csv")
all_Zplusjets = glob.glob(path + "/Z_plus_jets_*.csv")
```

```python
li_Signal = []
li_SingleTop = []
li_ttbar = []
li_ttbarV = []
li_WW = []
li_ZZ = []
li_ZW = []
li_Zplusjets = []

#Signal
for filename in all_Signal:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_Signal.append(df)

Signal = pd.concat(li_Signal, axis=0, ignore_index=True)

#Single Top
for filename in all_SingleTop:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_SingleTop.append(df)

SingleTop = pd.concat(li_SingleTop, axis=0, ignore_index=True)

#ttbar
for filename in all_ttbar:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_ttbar.append(df)

ttbar = pd.concat(li_ttbar, axis=0, ignore_index=True)

#ttbarV
for filename in all_ttbarV:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_ttbarV.append(df)

ttbarV = pd.concat(li_ttbarV, axis=0, ignore_index=True)

#WW
for filename in all_WW:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_WW.append(df)

WW = pd.concat(li_WW, axis=0, ignore_index=True)

#ZZ
for filename in all_ZZ:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_ZZ.append(df)

ZZ = pd.concat(li_ZZ, axis=0, ignore_index=True)

#ZW
for filename in all_ZW:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_ZW.append(df)

ZW = pd.concat(li_ZW, axis=0, ignore_index=True)

#Z plus jets
for filename in all_Zplusjets:
    df = pd.read_csv(filename, index_col=None, header=0)
    li_Zplusjets.append(df)

Z_plus_jets = pd.concat(li_Zplusjets, axis=0, ignore_index=True)
```

```python
# Concat diboson data
Diboson = pd.concat([WW, ZZ, ZW], axis=0, ignore_index = True)
Diboson
```

```python
List = [Diboson, Signal, SingleTop, ttbar, ttbarV, Z_plus_jets]
#Check for NaNs
for i in range(0,len(List)):
    print(i, List[i].isnull().values.any())
```
